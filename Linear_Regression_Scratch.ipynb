```python
import numpy as np
```

```python
#Linear Regression

class LinearRegression:
    def __init__(self, learning_rate=0.01, n_iter=1000):
        self.bias = None
        self.weights = None
        self.lr = learning_rate
        self.n_iter = n_iter

    def fit(self, X, y):
        # No. of rows, No. of features
        m, n = X.shape

        # Initializing with random vals
        self.bias = 0;
        self.weights = np.zeros(n) #[0, 0, .., 0]-> size == no. of features

        # Gradient Descent Procedure
        for i in range(self.n_iter):
            #The predicted output values (ŷ) for every training example
            y_pred = self.bias + np.dot(X, self.weights) 

            db =  (1/m) * np.sum(y_pred - y)  # Derivative with respect to bias
            dw = (1/m) * np.dot(X.T, (y_pred - y))  # Derivative with respect to θ or weights

            self.bias = self.bias - self.lr * db
            self.weights = self.weights - self.lr * dw

    def predict(self, X):
        #Predictions of y_test
        y_pred = self.bias + np.dot(X, self.weights)
        return y_pred
```

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

model = LinearRegression()

model.fit(X, y)

y_pred = model.predict(X)
print(y_pred)

print(model.bias)   #It should be near to 0 or 0
print(model.weights)  #It should be near to 2 or 2
```
